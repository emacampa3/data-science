1st method

```python
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # adds some axis
plt.show()
```

2nd method

```python
fig = plt.figure() # creates a figure
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y) # add some data
plt.show()
```

3rd method (recommended)

```python
fig, ax = plt.subplots()
ax.plot(x,y);

type(ax), type(fig)
```

##### Matplotlib example workflow

```python
# Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# Prepare data
x = [1,2,3,4]
y = [10,11,12,13]

# Setup plot
fig, ax = plt.subplots(figsize=(5,5)) # (width,height)

# Plot data
ax.plot(x,y)

# Customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

# Save & show 
# fig.savefig("images/sample-plot.png")
```
