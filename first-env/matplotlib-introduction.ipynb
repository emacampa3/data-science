##### Matplotlib example workflow

```python
# Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# Prepare data
x = [1,2,3,4]
y = [10,11,12,13]

# Setup plot
fig, ax = plt.subplots(figsize=(5,5)) # (width,height)

# Plot data
ax.plot(x,y)

# Customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

# Save & show 
# fig.savefig("images/sample-plot.png")
```

##### Making figures with NumPy arrays
* Line plot, Scatter plot, Bar plot, Histogram and Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100) 
# returns evenly spaced numbers over a specified interval: 100 numbers between 0 and 10
x[:50]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use the same data and make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
      ylabel="Price ($)");
```

```python
# Horizontal bar plot
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# Make some data and create a histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

##### Two options for subplots

Option 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10,5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(30), np.random.random(30));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

Option 2

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(30), np.random.random(30));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

##### Plotting from pandas Dataframes

```python
import pandas as pd
import numpy as np
```

```python
ts = pd.Series(np.random.randn(1000),
              index=pd.date_range("1/1/2020", periods =1000))
ts = ts.cumsum()
ts.plot();
```

```python
# Make a DataFrame
car_sales = pd.read_csv("car-sales.csv")
car_sales 
```

```python
# ReGex: replacing symbols from Price column with empty string
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales 
```

```python
# Removing last two zeros from the Price string using slicing
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
# Creating a new column with date_range
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

String to integer: `car_sales["Price"] = car_sales["Price"].astype(int)`

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Plotting the total sales
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
# Scatter plot
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```
