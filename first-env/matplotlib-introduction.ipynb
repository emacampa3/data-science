##### Making a plot from dictionary

```python
# Creating a dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
# Creating a bar graph, where the x-axis is the keys of the dictionary
# and the y-axis is the values of the dictionary
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())

# Customizing the plot
ax.set(title="Dan's Nut Butter Store",
      ylabel="Price ($)");
```

```python
# Horizontal bar plot
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# Making a random NumPy array of 1000 normally distributed numbers (x) and creating a histogram plot of x
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

##### Two options for subplots (figures with multiple plots on it)

Option 1

```python
# Creating an empty subplot with 2 rows and 2 columns (4 subplots)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10,5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(30), np.random.random(30));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

Option 2

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(30), np.random.random(30));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

##### Plotting from pandas Dataframes

```python
import pandas as pd
import numpy as np
```

```python
# Making a DataFrame
car_sales = pd.read_csv("car-sales.csv")
car_sales 
```

```python
# Creating a new column with date_range
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

Since `Price` column is not numeric data, it has to be changed into such using `regex`.

```python
# Replacing symbols from Price column with empty strings
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')

# Removing last two zeros from the Price string using slicing
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales

# Total Sales and Sale Date are now numeric
```

String to integer: `car_sales["Price"] = car_sales["Price"].astype(int)`

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Plotting the total sales
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
# Scatter plot
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

##### Which method to choose? PyPlot vs matplotlib OO method
* When plotting something quickly, PyPlot method is useful
* When plotting something advanced, OO method is better

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

Scatter plot with age and cholesterol using PyPlot method

```python
over_50.plot(kind="scatter", x="age",y="chol", c="target"); # c = colour
```

OO method mixed with PyPlot method

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
# changing the length of x-axis: ax.set_xlim([45, 100]) 
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customizing the plot
ax.set(title="Heart Disease and Cholesterol Levels", 
      xlabel="Age", 
      ylabel="Cholesterol")

# Adding a legend
ax.legend(*scatter.legend_elements(), title="Target")
# * grabs the values from scatter and unpacks the elements into legend

# Adding an average horizontal line across the axis
ax.axhline(over_50["chol"].mean(), linestyle="--", c="gray");
```

```python
# Subplot of chol, age and thalach: a single figure with two axes
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True) # both subplots have the same x value

# Adding a title to the whole Figure
fig.suptitle("Heart Disease Analysis", fontsize=14, fontweight="bold")

# Adding data to ax0
scatter = ax0.scatter(x=over_50["age"], 
                     y=over_50["chol"],
                     c=over_50["target"])
# Customizing ax0
ax0.set(title="Heart Disease and Cholesterol Levels", 
      ylabel="Cholesterol")

# Adding a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Adding a meanline
ax0.axhline(over_50["chol"].mean(), linestyle="--", c="gray");


# Adding data to ax1
scatter = ax1.scatter(x=over_50["age"], 
                     y=over_50["thalach"],
                     c=over_50["target"])
# Customizing ax1
ax1.set(title="Heart Disease and Max Heart Rate", 
      xlabel="Age", 
      ylabel="Max Heart Rate")

# Adding a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Adding a meanline to ax1
ax1.axhline(over_50["thalach"].mean(), linestyle="--", c="gray");
```

##### Customizing matplotlib plots

```python
# See the different styles
plt.style.available
# using a different style in whole notebook: plt.style.use("WANTED-STYLE")
```

```python
# Subplot of chol, age and thalach: a single figure with two axes
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True) # both subplots have the same x value

# Adding a title to the whole Figure
fig.suptitle("Heart Disease Analysis", fontsize=14, fontweight="bold")

# Adding data to ax0
scatter = ax0.scatter(x=over_50["age"], 
                     y=over_50["chol"],
                     c=over_50["target"], cmap="tab10")
# Customizing ax0
ax0.set(title="Heart Disease and Cholesterol Levels", 
      ylabel="Cholesterol")

# Changing the x-axis limits for both 
ax0.set_xlim([50, 80])

# Adding a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Adding a meanline
ax0.axhline(over_50["chol"].mean(), linestyle="--", c="gray");


# Adding data to ax1
scatter = ax1.scatter(x=over_50["age"], 
                     y=over_50["thalach"],
                     c=over_50["target"], cmap="tab10")

# Changing the y-axis limits
ax1.set_ylim([60, 200])

# Customizing ax1
ax1.set(title="Heart Disease and Max Heart Rate", 
      xlabel="Age", 
      ylabel="Max Heart Rate")

# Adding a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Adding a meanline to ax1
ax1.axhline(over_50["thalach"].mean(), linestyle="--", c="gray");
```
